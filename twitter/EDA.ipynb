```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("tweeter_data_pandas")

df.head()
```

```python
df["date"] = df["posted_time"].astype("datetime64")
df.groupby(df["date"].dt.hour).count().plot(kind="bar")
```

```python
df.hist(figsize=(6,6))
plt.tight_layout()
```
